# Credit Risk Analysis: Loan Default Prediction

**Project Type:** Supervised Machine Learning (Classification)  
**Domain:** Banking / Financial Services  

This project focuses on predicting loan default risk using applicant and loan-related attributes to support data-driven credit approval decisions.


## Problem Statement

Financial institutions face significant losses when borrowers fail to repay their loans. The objective of this project is to build a machine learning model that predicts the likelihood of loan default based on applicant and loan characteristics, enabling lenders to make informed, risk-aware lending decisions.


## What is Loan Default?

Loan default occurs when a borrower fails to repay the loan according to the agreed terms, either by missing scheduled payments or being unable to repay the full loan amount.

In practical terms:
- The borrower stops making payments
- The lender may incur financial loss
- Recovery is often costly and uncertain

From a data science perspective, this is framed as a **binary classification problem**:
- `1` → Default
- `0` → Fully paid / Non-default


## Why Is Predicting Loan Default Important?

Predicting loan default is critical because it directly impacts multiple aspects of a financial institution's operations.

### Financial Stability
- Defaulted loans reduce profitability
- High default rates increase overall credit risk exposure

### Better Lending Decisions
- Helps determine whether to approve, reject, or closely monitor applicants
- Enables risk-based interest rate pricing

### Operational Efficiency
- Reduces reliance on manual credit assessment
- Supports automated and scalable decision-making systems

### Regulatory & Risk Management
- Financial institutions must justify lending decisions
- Explainable and transparent models are essential for regulatory compliance

**Key takeaway:**  
> Accurate default prediction balances profitability with risk, helping lenders lend responsibly.


## Cost of False Positives vs False Negatives

In credit risk modeling, different types of prediction errors have different business impacts.

### False Negative (Most Costly)
**Definition:**  
The model predicts a borrower as non-default, but the borrower actually defaults.

**Impact:**
- Risky loan is approved
- Direct financial loss to the institution
- Additional collection and legal costs
- Increased risk in the loan portfolio

**Business cost:** High

---

### False Positive
**Definition:**  
The model predicts a borrower as default, but the borrower would have repaid the loan.

**Impact:**
- Lost business opportunity
- Potential customer dissatisfaction
- Minor revenue loss

**Business cost:** Lower than false negatives


## Business Conclusion

In credit risk modeling, minimizing **false negatives** is typically more important than minimizing false positives, even if it results in slightly lower loan approval rates.

This business requirement directly influences modeling choices, such as:
- Emphasizing **recall** over accuracy
- Analyzing precision–recall trade-offs
- Adjusting classification thresholds based on risk tolerance
